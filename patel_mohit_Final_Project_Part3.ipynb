# Visualization

Below is an interactive visualization I created from the dataset that will be usefull in gaining knowledge about which contry emitted what amount of Co2 in which specific year. 

```python
#importing libraries
import pandas as pd
import bqplot 
import numpy as np
import ipywidgets 
import matplotlib.pyplot as plt
import bqplot.pyplot
from IPython import display

#loading the data
df = pd.read_csv('co2_emissions_tonnes_per_person.csv')
```

```python
#This is the overview of the dataset
df
```

As we can see there are a lot of NaN values in the dataset. This can cause issues in creating analyzes. Let us firt replace the NaN with 0's. 

```python
#replacing NaN with 0's
df = df.fillna(0)
df
```

The dataset has 194 rows and 220 columns. Now the first column is called country which has the name of 193 different countries of the world. The rest of 219 columns represent years ranging from 1799 to 2017. Each row values represent the Co2 Emission per per in tonnes of each country throughout the given years. 
The command that we ran in the code block above was to replace all NaN with 0's and we can clealry see in the above code snippet that the O's are inplcae correctly

```python
#getting the column names using for-loop
x_axis = []
for columns in df.columns:
        x_axis.append(columns)

x_axis.pop(0)
# print(x_axis)

#Converting list to a np.array
np_x_axis = np.array(x_axis)
np_x_axis
```

```python
#Getting the country column
y_axs = df['country']
y_axs
```

# Let's Get INTERACTIVE !!!

```python
#creating ipywidgets 
@ipywidgets.interact(x_axis = np_x_axis,
                     y_axis = y_axs)

#Let's create a function to make the graphical visualization.

def plot1(x_axis, y_axis):
    fig,ax = plt.subplots()
    plt.rcParams["figure.figsize"] = (20,5)
    ax.scatter(x=x_axis, y=df[x_axis][df.index[df['country']==y_axis].tolist()])
    #new_first10.index[new_first10['country']==y_axis].tolist() this code is used to get specif index of a row
    #that matches a column value.
    #reference:- https://www.statology.org/pandas-get-index-of-row/
    plt.ylabel('Co2 Emission in tons')
    plt.xlabel('Year')
```

# You must be shouting, "Too many years to Select"

Let's make this a little easy for you!!!

```python
#We can easily spot that until the year 1950 almost all countries had no Co2 Emission. We can get rid of those elements
x_axs = x_axis.copy()
new_x_ax = x_axs[150:]
```

```python


#creating ipywidgets 
@ipywidgets.interact(y_axis = y_axs)

#Let's create a function to make the graphical visualization.

def plot2(y_axis):
    fig,ax = plt.subplots()
    plt.rcParams["figure.figsize"] = (20,5)
    ax.scatter(x=new_x_ax, y=df[new_x_ax][df['country']==y_axis])
    #new_first10.index[new_first10['country']==y_axis].tolist() this code is used to get specif index of a row
    #that matches a column value.
    #reference:- https://www.statology.org/pandas-get-index-of-row/
    plt.ylabel('Co2 Emission in tons')
    plt.xlabel('Year')
    plt.xticks(rotation = 90)
    plt.show()
    
```

# What is Happening to the Graphs ???

So you must be thinking what is happening in the above visualizations. The first visualization is created for you to select what country you want to investigate for Co2 Emission. You can pick you x-axis by choosing the year you like and you can then pick the Country you want to select. The graph will give out the value of the Co2 Emission of that country in that year. 

I used ipywidgets to generate 2 widgets that allows you to make a selection of fields. After the field is selected the function plot 1 and plot 2 are loading that specific selected fields and generating the x and y axis of the plot. After that the plt commands are setting up the sizes,  labels, and fonts of the graphs. Ultimately the graph is created and is getting updated every time you decide to change the field. 

To make it more easier for you the second visualization has a little more cleaned data, meaning all the unimportant data where 99.9% of the countries had no Co2 Emission has been removed. We don't want to see 0 values in our graphs. The second visualization has also selected all the years from 1949 to 2017 so you can actually see the pattern on Co2 Emission. This makes it easy to see which countries has constanlt increased or decreased their Co2 Emission. You still have the option to pick the country of you liking :)


# Let's try a different one!!

It is always good practice to take fraction of the dataset and work our code to create visualization in the start. The reason is when dealing with smaller sample size data we can easily find error and edit the code easily. The runtime also decreases which makes our work more efficient. So we will also first take 10 columns and 10 rows worth of data to create an interactive dashboard, and then once we get that running we can load the entire dataset. 

```python
df1 = df[df.columns[range(1)]] #getting first column
df2 = df.iloc[:,-10:] #getting second column
new = df1.join(df2) #joining 2 dataframes
new
```

```python
#new_first10 is the dataframe containing 10 rows and columns from the original dataframe
new_first10 = new.head(10) 
new_first10
```

```python
#Creating a pivot table for the sample dataframe
df_pivot = pd.pivot_table(new_first10, index = 'country')
```

```python
#Checking the index
df_pivot.index
```

```python
#Checking the columns
df_pivot.columns
```

```python
#Checking the values
df_pivot.values
```

```python
#Here we are setting up of barplot graph
#scales
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale() 

#axis
x_ax = bqplot.Axis(label='Years', scale=x_sc)
y_ax = bqplot.Axis(label='Co2 Emission', scale=y_sc, 
                   orientation = 'vertical')


#marks
bar_plot = bqplot.pyplot.bar(x=[], y=[], scales={'x':x_sc, 'y':y_sc},padding=0.4)
#We will fill the blank x,y parameter in the on_selected function

#figure
fig_bar_plot = bqplot.Figure(marks=[bar_plot], axes=[x_ax,y_ax])
```

```python
#Lets plot the data using a gridheatmap

#Scales
col_sc = bqplot.ColorScale()
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()


#Axis
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')

x_ax = bqplot.Axis(scale = x_sc, label='Years')
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical', 
                   label = 'Country')

# Marks
heat_map = bqplot.GridHeatMap(color = df_pivot.values, #Each block represents the Co2 Emission value
                              row=df_pivot.index, #The country names
                              column = df_pivot.columns , #The Years
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              
                              interactions = {'click': 'select'},
                              selected_style = {'fill':'blue'})
```

```python
# interactivity 
myLabel = ipywidgets.Label()
def on_selection(change):
    if len(change['owner'].selected) == 1: #only 1 selected
        i,j = change['owner'].selected[0]
        v = df_pivot.values[i,j] # grab data value
        myLabel.value = 'Co2 Emission:- ' + str(v) 

#         Here first I tried to directly make a scatter plot but it didn't work the way I intended
#         I realized to make that change I would have to make a bqplot.scatter instead of bqplot.bar
#         fig,ax = plt.subplots()
#         ax.scatter(x=df_pivot.columns[j], y=df_pivot.values[i,j])
#         plt.show()

        bar_plot.x = df_pivot.columns[[j]] #It is grabbing the column name(year) that we select from the heatmap
        bar_plot.y = df_pivot.values[i,[j]]#We are grabbing the value of Co2 Emission of a specific country during 
                                           # a specific year
        bar_plot.colors='green' # Setting the color of the bars
    
#observe will help us check our data     
heat_map.observe(on_selection, 'selected')

```

```python
# Figure
fig = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig.layout.min_width = '700px'
fig_bar_plot.layout.min_width = '500px'

myDashboard = ipywidgets.HBox([fig,fig_bar_plot])
myDashboard = ipywidgets.VBox([myLabel,myDashboard])

myDashboard   
```

# Summary of Dashboard

For the visualization we first created a heatmap then linked it with a bar plot using bqplot. As mentioned above each block in the gridheatmap represents the data of specific row and columns value. When the user selects a code block the machines goes through the pivot table and finds the specific year selected. Then it goes through the country columns to find the specific country and then grabs the row,column value or the Co2 Emission value of that particular country in that particular year. Then a bar plot is generated on the right which represents the Co2 Emission value. There is only one bar for each block because it  contains only one value and the dataset also doesn't have any relative columns that we can groupby and create some other visualization. 

Overall the visualization will tell the user of what Co2 Emission(per person in tonnes) is given out in atmosphere by the selected country and the selected year. This is just a sample dashboard presented because loading the entire dataset will cause runtime error or crash the visualization as it takes up a lot of memory.
